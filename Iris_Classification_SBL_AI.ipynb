```python
import sklearn
import numpy as np
import pandas as pd
from sklearn import datasets
```

### 2. Data Understanding

```python
iris = datasets.load_iris()
```

```python
iris    # data -> store with dictionary
```

```python
iris_data = pd.DataFrame(iris.data, columns = ['Sepal Length', 'Sepal Width', 'Petal Length', 'Petal Width'])
```

```python
iris_data
```

```python
iris_data['Class'] = iris.target
```

### 3. Data Preprocessing

```python
iris_data.dropna(axis=0, inplace=True)   # Remove row of'NaN' value in real data set
```

```python
X= iris_data.iloc[:, 0:-1]     # separate row and col , desire one is at last part
Y= iris_data.iloc[:, -1]
```

```python
# If there is not separated dataset --> Separate training and testing dataset

from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 3)   # testing = 20% , training = 80%
```

```python
X_train
```

### 4. Modeling

```python
from sklearn.linear_model import LogisticRegression    # LogisticRegression for classification
model = LogisticRegression(max_iter=1000)
model.fit(X_train, Y_train)
```

### 5. Evaluation

```python
y_predict = model.predict(X_test)  # testing for not traning data giving features (X)
```

```python
y_predict
```

```python
from sklearn.metrics import accuracy_score, precision_score, f1_score, recall_score
print('Accuracy Score',accuracy_score(Y_test, y_predict)*100,'%')
print('Precision Macro Score',precision_score(Y_test, y_predict,average = 'macro')*100,'%')
print('Recall_Score',recall_score(Y_test, y_predict, average = 'macro')*100,'%')
print('F_Score',f1_score(Y_test, y_predict, average = 'macro')*100,'%')
```
